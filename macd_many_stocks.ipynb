```python
from mod_my_utils.read_write_data import get_price_yahoo
```

```python
from mod_macd_utils.macd import get_macd_signal_line, get_signal_macd_crossover
from mod_macd_utils.macd import get_signal_macd_signalline, plot_macd_buy_sell
```

```python
symbols = ["HDFC", 'RELIANCE', 'TCS', 'HINDUNILVR', 'INFY', 'HDFC', 'BHARTIARTL', 'ITC', 'ASIANPAINT', 'NESTLEIND', 'HCLTECH', 'SBI', 'SBIN']
```

```python
### stock parameters
data_source = "yahoo"
# MM/DD/YYYY
start_date = "1/1/2010"
end_date = "1/1/2015"
datasource = data_source
print(start_date)
print(end_date)
print(datasource)
```

```python
# macd parameters
short_window = 20
long_window = 50
signal_window = 12
```

```python
for symbol in symbols[:5]:
    print(symbol)
    data = get_price_yahoo(symbol, data_source, start_date, end_date)
    data = data[["Adj Close"]].copy()
    data.columns = ["price"]
    macd_data = get_macd_signal_line(data, short_window, long_window, signal_window)
    macd_signals_from_mcad = get_signal_macd_crossover(macd_data, long_window)
    macd_signals_from_signalline = get_signal_macd_signalline(macd_data, long_window)
    print("Signals from macd cross over:")
    plot_macd_buy_sell(macd_data, macd_signals_from_mcad, symbol = symbol)
    print("Signals from macd and signal line")
    plot_macd_buy_sell(macd_data, macd_signals_from_signalline, symbol = symbol)
```
